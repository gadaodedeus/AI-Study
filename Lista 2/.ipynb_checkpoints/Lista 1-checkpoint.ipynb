```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
#import plotly.express as px

df=pd.read_csv('dados_banco.csv')
```

```python
Exercicio 1. media, mediana, desvio padrao
```

```python
print('Media: '+str(df['Idade'].mean())+'\nMediana: '+str(df['Idade'].median())+'\nDesvio Padrão: '+str(df['Idade'].std()))
```

```python
Exercicio 2. liste colunas com null e preencha c alguma estrategia
```

```python
cols_p_converter = ['Saldo_cc', 'Saldo_poupanÃ§a', 'Saldo_investimento', 'Devedor_cartao']
for col in cols_p_converter:
    df[col] = df[col].str.replace(',', '.').astype('float64')
print('Not a number:\n', df.isna().sum())
```

```python
df=df.fillna(0)
print('Not a number:\n', df.isna().sum())
```

```python
OBS: saldo da conta nao é uma variavel que poderia ser preenchida com uma media de todos os saldos, entao podemos assumir que esta conta nao tem dinheiro
```

```python
Exercicio 7. normalization
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for col in df.columns:
    if df[col].dtypes == 'object':
        df[col] = le.fit_transform(df[col])
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X_norm = mms.fit_transform(df)
print(X_norm)
x, y = X_norm.min(axis=0), X_norm.max(axis=0)
print(x)
print(y)
```
