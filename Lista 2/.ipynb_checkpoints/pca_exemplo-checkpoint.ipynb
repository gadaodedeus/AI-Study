## Análise dos Componentes Principais (PCA)

```python
#Importando bibliotecas clássicas do Python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

#Montando a partição do Google Drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/IA_2023/PCA/iris.csv'

#Abrindo um arquivo (tabular) do tipo csv
df = pd.read_csv(file_path, delimiter=",")

#Filtando as classes existentes
classes = np.unique(df[df.columns[-1]])
#print(classes)

#Pega a lista de atributos
attribute_list = list(df.columns)
print(attribute_list)

df
```

Construir as variáveis $X$ e $Y$, sendo que $X$ será composta pelo atributos e $Y$ as classes.

```python
df = df.values #gera um ndarray (numpy)
nrow, ncol = df.shape
X = df[:,0:ncol-1]
Y = df[:,-1]
```

Padronização dos atributos para torná-los na mesma escala numérica.

```python
from sklearn.preprocessing import StandardScaler #Padronização tipo N(0,1)
scaler = StandardScaler().fit(X)
X = scaler.transform(X)
```

Análise dos componentes principais (redução de dimensionalidade via PCA).

```python
from sklearn.decomposition import PCA
#Seta n=2 como o número de componentes principais
pca = PCA(n_components=3)
#Cálculo do PCA
pca_result = pca.fit_transform(X)
```

Visualizaçao dos dados projetados em duas (n=2) dimensões.

```python
plt.figure(figsize=(8,5))
plt.scatter(pca_result[:,0], pca_result[:,1], pca_result[:,2], s=50, color = 'blue')
plt.xlabel("Componente 1", fontsize=20)
plt.ylabel("Componente 2", fontsize=20)
plt.ylabel("Componente 3", fontsize=20)
plt.xticks(color='k', size=20)
plt.yticks(color='k', size=20)
#plt.zticks(color='k', size=20)
plt.show(True)
```

Colorindo de acordo com as classes de cada instância.

```python
#Define uma paleta de cores
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']
aux = 0
plt.figure(figsize=(8,5))
for c in classes:
    nodes = np.where(Y == c)
    # plota os dados, considerando uma cor para cada classe
    plt.scatter(pca_result[nodes,0], pca_result[nodes,1], s=50, color = colors[aux], label = c)
    aux = aux + 1
plt.legend()
plt.xlabel("Componente 1", fontsize=20)
plt.ylabel("Componente 2", fontsize=20)
plt.xticks(color='k', size=20)
plt.yticks(color='k', size=20)
plt.show(True)
```

Estimatindo o número de componentes no PCA para projetar os dados via análise da variância.

```python
pca = PCA().fit(X)
plt.figure(figsize=(8,5))
ncomp = np.arange(1, np.shape(X)[1]+1)
# mostra a variância acumulada
plt.plot(ncomp, np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.xlabel('Número de componentes', fontsize=20)
plt.ylabel('Variância explicada acumulativa', fontsize=20);
plt.xticks(color='k', size=20)
plt.yticks(color='k', size=20)
plt.grid(True)
plt.show(True)
```
