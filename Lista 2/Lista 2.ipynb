6. Considere os dados do DataFrame “iris.csv”. Calcule a média, variância e IQR para cada
atributo.

```python
for i in df:
  np.sort(df[i])
  print(f'Média {i}: {df[i].mean()}')
  print(f'Variância {i}: {df[i].var()}')
  Q1 = np.percentile(df[i], 25)
  Q3 = np.percentile(df[i], 75)
  IQR = Q3 - Q1
  print(f'IQR {i}: {IQR}\n')
```

7. Calcule as correlações de Pearson e Spearman entre as colunas ‘Circ’ e ‘Comp’ do Data-
Frame “vehicle.csv”

```python
file_path = '/content/drive/MyDrive/Inteligência Artificial/vehicle.csv'
df = pd.read_csv(file_path, delimiter=",")
```

```python
from scipy.stats import spearmanr

corr_pearson = df['Circ'].corr(df['Comp'])
corr_spearman, _ = spearmanr(df['Circ'], df['Comp'])

print(f'Correlação de Pearson: {corr_pearson}')
print(f'Correlação de Spearman: {corr_spearman}')
```
